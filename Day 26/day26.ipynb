```python
#Exercise 2
list_of_strings = input().split(',')
# 🚨 Do  not change the code above

# TODO: Use list comprehension to convert the strings to integers 👇:
int_list = [int(x) for x in list_of_strings]

# TODO: Use list comprehension to filter out the odd numbers
# and store the even numbers in a list called "result"
result = [x for x in int_list if x%2==0 ]

# Write your code 👆 above:
print(result)
```

```python
#Exercise 3
with open("file1.txt") as file:
  file1 = file.read()
file1 = file1.split('\n')
file1 = [int(x) for x in file1 if x!='']

with open("file2.txt") as file:
  file2 = file.read()
file2 = file2.split('\n')
file2 = [int(x) for x in file2 if x!='']

result = [x for x in file1 if x in file2]


# Write your code above 👆
print(result)
```

```python
# Exercise 1
sentence = input()
# 🚨 Don't change code above 👆
# Write your code below 👇

sentence = sentence.split(' ')
result = {word:len(word) for word in sentence}

print(result)
```

```python
weather_c = eval(input())
#The eval() function will help us convert the string input into a Python dictionary, provided that the inputs are already formatted with the correct syntax. 
# 🚨 Don't change code above 👆

# Write your code 👇 below:
weather_f = { key:(value*9/5)+32 for (key,value) in weather_c.items()}

print(weather_f)
```
