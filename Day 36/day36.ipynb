```python
import requests
STOCK_NAME = "TSLA"
COMPANY_NAME = "Tesla Inc"
key = "FUH4QULX9JEESAN7"
params = {
    "apikey":key,
    "function": "TIME_SERIES_DAILY",
    "symbol":STOCK_NAME

}


STOCK_ENDPOINT = "https://www.alphavantage.co/query"
NEWS_ENDPOINT = "https://newsapi.org/v2/everything"

response = requests.get(STOCK_ENDPOINT,params=params)
```

```python
data =  response.json()
data = data['Time Series (Daily)']
```

```python
import datetime
if datetime.datetime.now().weekday()==6:
    yesterday =   str(datetime.datetime.now().date() - datetime.timedelta(days=2))
    yesterday2 = str(datetime.datetime.now().date() - datetime.timedelta(days=3))
else:
    yesterday =   str(datetime.datetime.now().date() - datetime.timedelta(days=1))
    yesterday2 = str(datetime.datetime.now().date() - datetime.timedelta(days=2))

stock_values = [float(value['4. close']) for key,value in data.items() if key in [yesterday,yesterday2]]
stock_values
```

```python
stock_diff = abs(stock_values[0]-stock_values[1])
```

```python
stock_diff = ((stock_diff*100)/stock_values[1])
```

```python
if stock_diff>5:
    print("Get News")
```
